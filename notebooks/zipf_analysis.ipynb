# Section 1

  * here we read the text (both of the texts) then we split them into words

```python
# read Text-1.txt file
with open('Text-1.txt', 'r', encoding='ISO-8859-1') as f: text1 = f.read()

# first show the number of characters in the file
print('Number of characters in Text-1.txt file:', len(text1))

# second show the number of words in the file
words = text1.split()
print('Number of words in Text-1.txt file:', len(words))

# convert all words to lower case
words = [word.lower() for word in words]

# third show the number of lines in the file
lines = text1.split('\n')
print('Number of lines in Text-1.txt file:', len(lines))
```

```python
# read Text-2.txt file
with open('Text-2.txt', 'r', encoding='ISO-8859-1') as f: text2 = f.read()

# first show the number of characters in the file
print('Number of characters in Text-2.txt file:', len(text2))

# second show the number of words in the file
words = text2.split()
print('Number of words in Text-2.txt file:', len(words))

# third show the number of lines in the file
lines = text2.split('\n')
print('Number of lines in Text-2.txt file:', len(lines))
```

# Section 2
  * Here we count each word in word list then save it to dict.

```python
# this is the list of all words that we want to count
words_to_count = [
  'the',
  'eye',
  'of',
  'existence',
  'hair',
  'and',
  'nature',
  'month',
  'head',
  'darwin',
  'size',
  'touch',
  'second',
  'conclusion',
  'revolution',
  'muscles',
  'necessity',
  'public',
  'body',
  'Within',
  'to',
  'in',
  'distribution',
  'words',
]

# count the number of words in Text-1.txt file
text1_count = {word: text1.count(word) for word in words_to_count}
print(text1_count)

# count the number of 'the' in Text-2.txt file
text2_count = {word: text2.count(word) for word in words_to_count}
print(text2_count)
```

# Section 3

# Section 5
  * Here we plot the sorted value to compare it to Zipf's law 

```python
# part B) show the word graph

# for Text-2.txt file with x as rank and y as count
# create a figure
fig, ax = plt.subplots(figsize=(15, 5))

# title of the figure
ax.set_title('Word Graph for Text-2.txt file')

# sort the DataFrame by rank
text2_df = text2_df.sort_values(by='rank')

# plot the bar chart for Text-2.txt file
ax.plot(text2_df['rank'], text2_df['count'])
# ax.bar(text2_df['rank'], text2_df['count'])

# log scale for y-axis
ax.set_yscale('log')
# set labels for y-axis
ax.set_ylabel('Count')

# log scale for x-axis
ax.set_xscale('log')
# set labels for x-axis
ax.set_xlabel('Rank')

# show the figure
plt.show()
```

```python
# zipf's law frequency = 1/(rank + b)^a
freq = lambda rank, a, b: 1 / (rank + b) ** a
# real frequency = count / total number of words
real_freq = lambda count, total: count / total

# calculate the frequency for Text-2.txt file
text2_df['freq'] = freq(text2_df['rank'], 3, 2.7) # a = 3, b = 2.7
# calculate the real frequency for Text-2.txt file
text2_df['real_freq'] = real_freq(text2_df['count'], len(words))

# plot the frequency and real frequency
fig, ax = plt.subplots(figsize=(15, 5))

# plot the frequency
ax.plot(text2_df['rank'], text2_df['freq'], label='Zipf\'s law')
# plot the real frequency
ax.plot(text2_df['rank'], text2_df['real_freq'], label='Real frequency')

# set labels for x-axis and y-axis
ax.set_xlabel('Rank')
ax.set_ylabel('Frequency')

# set the title of the figure
ax.set_title('Zipf\'s law vs Real frequency')

# log scale for x-axis and y-axis
ax.set_xscale('log')
ax.set_yscale('log')

# show the legend
ax.legend()

# show the figure
plt.show()
```

## Part B) Zipf's Law graph
***

First we have Zipf's Law formula:

$$f = \frac{1}{(r + b)^a}$$

Then we have the real frequency formula:

$$f = \frac{count}{total}$$

According to the Zipf's Law, the frequency of a word is inversely proportional to its rank. This means that the most frequent word will occur twice as often as the second most frequent word, three times as often as the third most frequent word, and so on. This is clearly visible in the graph, where the frequency of the most frequent word is much higher than the frequency of the second most frequent word, and so on.

Then we have this two parameters $a$ and $b$, which are used to adjust the curve of the graph. The value of $a$ is usually between 1 and 2, and the value of $b$ is usually between 2 and 3. The value of $a$ determines the slope of the curve, and the value of $b$ determines the position of the curve. And now I set $a = 3$ and $b = 2.7$. to better fit the data.
